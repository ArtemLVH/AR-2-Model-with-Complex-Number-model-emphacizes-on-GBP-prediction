```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the complex number representation
A = 100  # Amplitude of the cyclic pattern (represents the strength or intensity of a wave)
omega = 0.1  # Angular frequency determining the speed of oscillation.
phi = np.pi / 4   # Phase angle, indicating the timing of the cycle. Here, it's set to π/4 radians.* (the code below explains it)

# Generating time series data using the AR(2) model with complex numbers
num_periods = 20
time = np.arange(num_periods) #represents the number of time periods
GDP = A * np.cos(omega * time + phi) + np.random.normal(0, 10, num_periods)  # Adding some random noise, time array from np
#GDP array and random noise added.

# Plotting the original cyclic pattern
plt.figure(figsize=(10, 6))
plt.plot(time, GDP, label='Original GDP (with noise)', marker='o')
plt.title('Original GDP with Cyclic Pattern')
plt.xlabel('Time')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)
plt.show()
```

In this example, GDP represents the economic data with a cyclic pattern represented by a cosine function, and some random noise is added to simulate real-world variability. **The complex number, specifically the phase angle  ϕ, influences where the cyclic pattern starts.
By adjusting the phase angle, you determine at which point in the oscillation cycle the data begins, influencing how the GDP values evolve over time.**

You would fit the AR(2) model to actual GDP data to obtain meaningful parameter values and make more accurate predictions. The graph above serves to demonstrate the impact of the cyclic pattern represented by a complex number in the time series data.
In your specific AR(2) example with complex numbers influencing the cyclic pattern in GDP data, the autoregressive coefficients (ϕ1 and ϕ2) are related to the complex number used to model the cyclic behavior.


**Imagine you have a magical crystal ball that can predict how high a bouncing ball will go. This bouncing ball represents the economy, going up and down.
Now, the crystal ball has two magical numbers: one is a secret friend named Magic Multiplier 1 (let's call it M1), and the other is another secret friend named Magic Multiplier 2 (let's call it M2).**


M1 is like a guide telling us, "Today's bounce is a bit like yesterday's bounce, but maybe a little stronger or weaker."
M2 is another guide saying, "Today's bounce is also a bit like the bounce from the day before yesterday, but with its own touch."
So, if we want to predict today's bounce (Yt), we can use our crystal ball and its magical friends like this:

***-> Yt=M1⋅Yt−1+M2⋅Yt−2+(a sprinkle of magic randomness).***

The "sprinkle of magic randomness" is like saying there might be some unexpected surprises or changes that the crystal ball can't predict.
Now, how do we get these magical friends M1 and M2? Well, we use a special way of understanding the bouncing pattern with a magical complex number. This complex number has two parts: one part helps us know how high the ball likes to bounce (we call it amplitude), and the other part helps us know when the ball starts bouncing (we call it phase angle).
So, in our crystal ball language, M1 and M2 are like using this special complex number to guide us in predicting how the economy bounces up and down in a cool pattern. They are like the magical friends who help us make sense of the bouncing game.

To develop what we just said, let's create another simplified example where the complex number's impact is more visible. We'll use a sine wave, but the principles apply to cosine waves and more complex modeling scenarios:
Its very important to note mathematically this: 


If we don't use complex numbers, we could have uses separate cosine and sine functions:
y(t)=Acos(ϕ)sin(ωt)+Asin(ϕ)cos(ωt).

**BUT Using complex numbers, Euler's formula allows us to represent the sine wave more elegantly:
y(t)=A⋅Re{e^(j(ωt+ϕ))}
Here, j is the imaginary unit (j^2=−1).**
The complex representation consolidates the sine and cosine terms into a single expression, making it more compact and mathematically convenient.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 100  # Amplitude
omega = 0.1  # Angular frequency
phi_values = [0, np.pi/4, np.pi/2]  # Different phase angles

# Generating time series data using the sine wave
num_periods = 20
time = np.arange(num_periods)

# Plotting the sine wave with different phase angles
plt.figure(figsize=(12, 6))

for phi in phi_values:
    sine_wave = A * np.sin(omega * time + phi)
    plt.plot(time, sine_wave, label=f'Phase Angle = {phi:.2f}')

plt.title('Sine Waves with Different Phase Angles')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

When we change this special number (the phase angle of the complex number), it's like starting our secret code from a different point.
For example, with one special number, the GDP might start going up right from the beginning. With another special number, it might start going up a bit later.
So, changing this special number changes where our ups and downs begin in the graph.
Think of it like telling a story where the exciting parts happen at different times. The complex number helps us decide when the story of GDP starts getting exciting, and that's why changing it changes where the ups and downs begin in our graph. 

Let's get a bit more specific about what is a phase angle below in order to illustrate how we see complex numbers in our time series graphs

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1  # Amplitude
omega = 1  # Angular frequency (for simplicity)
phi = np.pi / 4  # Phase angle

# Generate time values
t = np.linspace(0, 2 * np.pi, 1000)  # One complete cycle

# Sine wave equation with phase angle
y = A * np.sin(omega * t + phi)

# Plot the sine wave
plt.figure(figsize=(8, 4))
plt.plot(t, y, label=f'Sine Wave with Phase Angle = {phi:.2f}', color='blue')
plt.title('Sine Wave with Phase Angle')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Horizontal line at y=0
plt.axvline(phi, color='red', linestyle='--', label='Phase Angle')  # Vertical line at phase angle
plt.legend()
plt.grid(True)
plt.show()
```

**REAL LIFE STUDY CASE Ocean Waves and Tides.**

Imagine you are studying the motion of ocean waves, particularly the periodic rise and fall of water. This motion can be modeled by a sine wave, and the phase angle could represent the initial position of the wave cycle.

Ocean waves exhibit a repeating pattern as they move towards the shore.
A phase angle of pi/4 could mean that, at a specific moment in time (e.g., the start of a day), the water level is already a quarter of the way through its typical oscillation cycle. Meaning : The waves go up and down, up and down, like a big, slow dance. It means the waves have covered 25% of the distance from their lowest point to their highest point. If we imagine the full cycle of a wave as going from the lowest point (like the bottom of a wave) to the highest point (like the crest of a wave), then "a quarter of the way" means we're capturing the wave when it's just started going up but hasn't reached its highest point yet. It's like catching the wave in the early part of its upward motion.


In simpler terms, if you were standing on the shore at sunrise (time t=0), the water might be starting its rise, a quarter of the way up the wave.
